```python
l=list(df.columns)
c=[] #categorical
b=[] #binary
iv=[] #interval
o=[] #ordinal
for i in l:
  if len(i)>2:
    if i[-3:]=="cat":
      c.append(i)
    elif i[-3:]=="bin":
      b.append(i)
    elif df[i].dtype == float:
      iv.append(i)
    elif df[i].dtype == int:
      o.append(i)
o.remove("target")
len(c),len(b)
```

##Handling missing values

*   For interval values replace -1 with mean
*   For other values replace -1 with median
*   ps_car_03_cat has median -1 so in this case -1 is repaced by 2



```python
l=df.columns[(df==-1).sum()>0]
df["ps_car_03_cat"]=df["ps_car_03_cat"].replace(-1,2)
for i in l:
  if i in iv:
    df[i]=df[i].replace(-1,df[i].mean())
  else:
    df[i]=df[i].replace(-1,(df[i].median()))
(df==-1).sum().sort_values()
```
